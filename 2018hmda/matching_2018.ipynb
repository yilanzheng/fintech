```python
import pandas as pd
```

```python
import fuzzywuzzy
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
panel18=pd.read_excel("2018panel.xlsx")
panel18.head()
```

```python
id17=pd.read_excel("match_2017.xlsx")
id17["name_clean"]=[i.upper().replace("NATIONAL ASSOCIATION","").replace(" & "," AND ").replace("&"," AND ").replace(",","").replace(".","").replace(" INC","").replace(" LLC","").replace(" THE","").replace(" COMPANY","").replace(" CORPORATION","").replace(" CORP","").replace(" CO","").replace(" NA","") for i in list(id17.iloc[:,0])]
id17.head()
```

```python
panel18["name_clean"]=[i.upper().replace(" & "," AND ").replace("&"," AND ").replace("NATIONAL ASSOCIATION","").replace(",","").replace(".","").replace(" INC","").replace(" LLC","").replace(" THE","").replace(" COMPANY","").replace(" CORPORATION","").replace(" CORP","").replace(" CO","").replace(" NA","") for i in list(panel18.iloc[:,1])]
panel18.head()
```

```python
panel18[panel18["respondent_name"]=="SUNTRUST BANKS, INC."].iloc[0,2]
```

```python
panel18=panel18.drop([0])
panel18.head()
```

```python
n=panel18.shape[0]
results=pd.DataFrame(columns=["lei","name18","name17","score"])

for i in range(0,n):
    input_name=panel18.iloc[i,2]
    match,score,temp=process.extractOne(input_name,id17.iloc[:,4],scorer=fuzz.ratio)
    name17=id17[id17["name_clean"]==match].iloc[0,0]
    results.loc[i]=(panel18.iloc[i,0],panel18.iloc[i,1],name17,score)
    print(i," ",panel18.iloc[i,2],"  ",name17,score)
```

```python
results.to_excel("results.xlsx")
```
